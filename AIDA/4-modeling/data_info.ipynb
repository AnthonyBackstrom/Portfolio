### Datan arvojen jakauman tutkiminen

Tässä on luotu kuvaajia datan arvojen jakauman tutkimiseksi, jotta voimme päättää käytämmekö keskiarvoa vai mediaania tyhjien arvojen täyttämiseen datatiedostossa.

Lopputulos:

* Hevosvoiman kanssa käytetään mediaania

* Kaupunkikulutuksen kanssa mediaania

* Myyjän luokituksen kanssa keskiarvoa (tai mediaania, ei niin väliä)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('your_file.csv', low_memory=False)

# Plot a histogram and a boxplot for the 'horsepower' column
df['horsepower'].plot(kind='hist', title='Horsepower Histogram')
plt.show()
df['horsepower'].plot(kind='box', title='Horsepower Boxplot')
plt.show()

# Plot a histogram and a boxplot for the 'city_fuel_economy' column
df['city_fuel_economy'].plot(kind='hist', title='City Fuel Economy Histogram')
plt.show()
df['city_fuel_economy'].plot(kind='box', title='City Fuel Economy Boxplot')
plt.show()

# Plot a histogram and a boxplot for the 'seller_rating' column
df['seller_rating'].plot(kind='hist', title='Seller Rating Histogram')
plt.show()
df['seller_rating'].plot(kind='box', title='Seller Rating Boxplot')
plt.show()

# Based on this, you can decide whether to use median or mean for imputation of missing values in each column
```
