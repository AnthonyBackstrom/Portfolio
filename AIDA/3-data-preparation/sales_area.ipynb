```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cleaned_df.csv')
```

```python
# Grouping data by region and counting the number of new and used cars
grouped = df.groupby(['dealer_region', 'is_new']).size().unstack().fillna(0)

# Building a bar chart
grouped.plot(kind='bar', stacked=False, width=0.4, position=0, figsize=(15, 10))
plt.title('Depending on region and machine type')
plt.xlabel('Dealer region')
plt.ylabel('Number of cars')
plt.xticks(rotation=0)  

plt.legend(title='Machine type', loc='upper right', labels=['New cars', 'Not new cars'])

plt.show()
```

```python
# Grouping data by region and counting the number of new and used cars
grouped1 = df.groupby(['dealer_region', 'body_type']).size().unstack().fillna(0)

# Building a bar chart
grouped1.plot(kind='bar', stacked=False, width=0.8, position=0, figsize=(15, 10))
plt.title('Dependence of region and body type')
plt.xlabel('Dealer region')
plt.ylabel('Number of cars')
plt.xticks(rotation=0)  

plt.legend(title='Body type', loc='upper right')

plt.show()
```

```python
# Grouping data by region and counting the number of new and used cars
grouped2 = df.groupby(['dealer_region', 'fuel_type']).size().unstack().fillna(0)

# Building a bar chart
grouped2.plot(kind='bar', stacked=False, width=1, position=0, figsize=(15, 10))
plt.title('Dependence of region and fuel type')
plt.xlabel('Dealer region')
plt.ylabel('Number of fuel type')
plt.xticks(rotation=0)  

plt.legend(title='Fuel type', loc='upper right')

plt.show()
```

```python
# Average horsepower for each region
average_horsepower = df.groupby('dealer_region')['horsepower'].mean()

# Plotting a graph of the relationship between region and horsepower
plt.figure(figsize=(15, 10)) 
average_horsepower.plot(kind='bar', color='blue')
plt.title('Dependence of the region and horsepower')
plt.xlabel('Dealer region')
plt.ylabel('Horsepower')
plt.xticks(rotation=0)
plt.show()
```

```python
# Grouping data by region and counting the number of new and used cars
grouped3 = df.groupby(['dealer_region', 'make_name']).size().unstack().fillna(0)

# Building a bar chart
grouped3.plot(kind='bar', stacked=False, width=0.8, position=0, figsize=(15, 10))
plt.title('Dependence of region and car brand')
plt.xlabel('Dealer region')
plt.ylabel('Number of cars')
plt.xticks(rotation=0) 

plt.legend(title='Car brand', loc='upper right')

plt.show()
```

```python
# Grouping data by region and counting the number of new and used cars
grouped4 = df.groupby(['dealer_region', 'transmission']).size().unstack().fillna(0)

# Building a bar chart
grouped4.plot(kind='bar', stacked=False, width=0.8, position=0, figsize=(15, 10))
plt.title('Dependence of region and car brand')
plt.xlabel('Dealer region')
plt.ylabel('Number of cars')
plt.xticks(rotation=0) 

plt.legend(title='Transmission', loc='upper right')

plt.show()
```

```python
# Average prices for each region
average_prices = df.groupby('dealer_region')['price'].mean()

# Plotting a graph of the relationship between region and average price
plt.figure(figsize=(15, 10)) 
average_prices.plot(kind='bar', color='blue')
plt.title('Dependence of the region and average car prices')
plt.xlabel('Dealer region')
plt.ylabel('Аverage car price')
plt.xticks(rotation=0)
plt.show()
```
